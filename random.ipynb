### Uniform distribution
***


```python
x = rng.uniform(low=0.0, high =10.0, size =10000)
x
```

```python
#round value in x is used to reduce the numbers
x_round = np.round(x, decimals = 0)

values, counts = np.unique(x_round,return_counts=True )

values, counts
```

```python
plt.bar(values,counts, edgecolor = 'Black')
```
